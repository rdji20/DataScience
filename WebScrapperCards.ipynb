````python
import requests
````

````python
from selenium import webdriver
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/usr/local/bin/chromedriver", chrome_options=options)
````

````python
# Robot clicking load more pages. We are going to store the page_sources in a page sources array

import time
from selenium.webdriver.support.wait import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC

page_sources = []

wait = WebDriverWait(driver, 5)
driver.get("https://everout.com/seattle/locations/?neighborhood=capitol-hill&neighborhood=first-hill&neighborhood=downtown-seattle&neighborhood=pioneer-square&neighborhood=central-district&neighborhood=eastlake&neighborhood=belltown&neighborhood=ballard&neighborhood=crown-hill&neighborhood=greenwood&neighborhood=phinney-ridge&neighborhood=fremont&neighborhood=madison-park&neighborhood=interbay&neighborhood=magnolia&neighborhood=south-lake-union&neighborhood=queen-anne&neighborhood=seattle-center&neighborhood=lower-queen-anne&neighborhood=west-queen-anne&neighborhood=east-queen-anne&neighborhood=wedgwood&neighborhood=ravenna&neighborhood=roosevelt&neighborhood=university-district&neighborhood=maple-leaf&neighborhood=sand-point&neighborhood=green-lake&neighborhood=wallingford&category=retail-and-hobbies&category=restaurants&category=bars&feature=walk-up-ordering&feature=takeout-covid-19&feature=delivery-covid-19&feature=curbside-pickup&feature=outdoor-dining&feature=covered-patio&feature=heated-patio&feature=black-owned&feature=livestreaming&feature=in-person-visits")
for x in range(39):
        page_source = driver.page_source
        page_sources.append(page_source)
        button = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[@class='page-link']")))
        if button.is_displayed():
            driver.execute_script("arguments[0].click();", button)
            time.sleep(2)
````

````python
naive_card_everout = {'title':[], 'description':[], 'where':[], 'linkImage':[]}
naive_card_everout = pd.DataFrame(naive_card_everout)
````

````python
import re

# Get the card information from all the links scrapped before

for link in links:
    try:
        htmlPage = urlopen(link) 
    except HTTPError as e: 
        print(e)
    except URLError as e:
        print('The server could not be found!')
    else:
        print("all good")
        time.sleep(2)
        url = link
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        header = soup.find("h1", class_="mb-0")
        title = str(header.string)
        print(title)
        address = soup.address
        if address is None :
            print("     no address")
        else :
            addressStr = str(address.get_text()).strip()
            addressStr = re.sub('\s+', ' ', addressStr)
            addressStr = addressStr.replace("website", "")
        website = "https://everout.com/seattle/"
        # Check descriptio
        description = soup.find("div", class_="description my-3")
        if description is None:
            print("        theres no descrp")
        else :
            description = str(description.string)
            description = re.sub('\s+', " ", description)
        image = soup.find("div", "item-image")
        if image is None:
            print("        no image")
        else :
            imageLink = image.img
        row = {'title':title, 'description':description, 'where':addressStr, 'linkImage':imageLink}
        naive_card_everout = naive_card_everout.append(row, ignore_index=True)
````
