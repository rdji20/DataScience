X1 = annual net sales/$1000

## Data Analysis
Data was already cleaned and transformed previously.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
greens_data = pd.read_excel("~/Downloads/LAB 2.10.4_ All greens franchise.xls")
```

```python
greens_data.shape, greens_data.head()
```

```python
corr_matrix = greens_data.corr()
```

```python
# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Create a linear regression model
lin_reg = LinearRegression()

# Split the data into X and y
y = greens_data.iloc[:, 0]  # First column
X = greens_data.iloc[:, 1:]  # All columns except the first one

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```

## Analysis Conclusion

The adjusted \( R^2 \) value is 0.997, which means that almost all the variance of the dependent variable is explained by our independent variables. Specifically, 99% of the variance is explained by:
- ( X2 ) = number of sq. ft./1000,
- ( X3 ) = inventory/\\$1000,
- ( X4 ) = amount spent on advertising/\\$1000,
- ( X5 ) = size of sales district/1000 families,
- ( X6 ) = number of competing stores in the district.

With the current data, we can infer that, holding all variables constant, for each additional competitor in the district, our annual net sales decrease by approximately \\$6,300. Furthermore, for every \\$1,000 we invest in advertising, keeping all other variables constant, we anticipate a return of about \\$10,600 throughout the year.

While our dataset is limited, it reveals a strong correlation between the selected variables and annual net sales. The logical progression would be to gather more data and also explore interaction effects. Specifically, it would be insightful to examine variations in the effectiveness of advertising expenditure, discerning if advertising yields better returns on investment at particular times of the year.
