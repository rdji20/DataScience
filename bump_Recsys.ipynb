# Creating LDA clusters

I will first use LDA to get the vectors and categories that the documents are going to be matching with.

```python
#
import string
porter = PorterStemmer()

words_to_skip = ["i", "canada", "seattle", "edmonton", "calgary", "vancouver", "toronto", "canadian", "edmonton", "alberta", "montreal", "sooner", '_', '...',"‘", '-', "'",'washington', "…", "’", "–", "like", "2022", "1", "10", "”", "“", "really", "new", "best", "favorite", "one", "amazing", "extraordinary", "super", "very", "inmaculate", "prettiest", "outstanding", "8", "5", "7", "10/10"]
stop_words = set(stopwords.words('english') + list(string.punctuation) + list(words_to_skip)) #Trying now without the tfidf + tfid_stop_words.term


texts = []
tokenizer = TweetTokenizer()
for d in curiocity_df.description:
    raw = d.lower()
    tokens = tokenizer.tokenize(raw)
    cleaned_tokens = []
    for token in tokens:
        if (token not in stop_words):
            cleaned_tokens.append(token)
    cleaned_tokens_filtered = []
    for ctoken in cleaned_tokens:
        if (ctoken not in words_to_skip):
            cleaned_tokens_filtered.append(ctoken)
    stemmed_tokens = []
    for token in cleaned_tokens_filtered:
        stemmed_tokens.append(porter.stem(token))
    texts.append(stemmed_tokens)
    
# Double stopwords to get rid of any derivate of the words that are not useful
# to categorize the documents e.g. (canada-man, token = canada, double filter
# for canada to be extracted)

double_stop_texts = []
for text in texts:
    doc = []
    for word in text:
        if word not in list(string.punctuation):
            if (word not in list(stop_words)):
                doc.append(word)
    double_stop_texts.append(doc)


dictionary = corpora.Dictionary(double_stop_texts)
corpus = [dictionary.doc2bow(text) for text in double_stop_texts]
lda_model = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=15)
index_best = similarities.MatrixSimilarity(corpus)
```

```python
# Step 2: comparing query vector to all seattle docs.

def compare_vector_with_docs_vectors(vector, docs_vectors):
    from scipy import spatial
    
    # Create the matrix for cosine similarity query
    mat_vector_query_2 = [0] * 15
    for tup in vector:
        mat_vector_query_2[tup[0]] = tup[1]
    mat_vector_query_2
    
    # Getting the matrix vector size 15 of all the seattle cards lda vectors
    count = 0
    recommended_cards_2 = []
    for card_vector in docs_vectors:
        mat_vector_sea = [0] * 15
        for tup in card_vector:
            mat_vector_sea[tup[0]] = tup[1]
        result = 1 - spatial.distance.cosine(mat_vector_query_2, mat_vector_sea)
        if result > 0.70 :
            pair = [count, result]
            recommended_cards_2.append(pair)
        count += 1
    return recommended_cards_2
```

```python
# Model to get recomended documents or places to go

# We have to define master graph and seattle database to call

# initiate this before >
def lda_query_docspace_vectorizer(query):
    import string
    import nltk
    from nltk.tokenize import RegexpTokenizer
    from nltk.corpus import stopwords
    from gensim import corpora, models, similarities
    from nltk.stem.porter import PorterStemmer
    from nltk.stem.snowball import SnowballStemmer
    from nltk.stem.lancaster import LancasterStemmer
    from nltk.tokenize import TweetTokenizer
    import os
    import tqdm
    
    stop_words_model = set(stopwords.words('english') + list(string.punctuation) + list(words_to_skip))

    texts2_model = []
    raw_model = query.lower()
    tokens_model = tokenizer.tokenize(raw_model)
    cleaned_tokens_model = []
    for token in tokens_model:
        if (token not in stop_words_model):
            cleaned_tokens_model.append(token)
    stemmed_tokens_model = []
    for token in cleaned_tokens_model:
        stemmed_tokens_model.append(porter.stem(token))
    texts2_model.append(stemmed_tokens_model)

    other_corpus_model = [dictionary.doc2bow(text) for text in texts2_model]
    unseen_doc_model = other_corpus_model[0]
    vector_model = lda_model[unseen_doc_model]

    # Check all the documents ids that are similar in the master data base
    # We are going to do cosine similarity with every documentsimi
    query_model = other_corpus_model[0]
    sims_model = index_best.get_similarities(query_model)
    count_model = 0
    docs_ids_user_query_model = []
    for i in sims_model:
        if i > 0.80:
            docs_ids_user_query_model.append(count_model)
        count_model += 1

    # Retrieve all documents (from specific location) which match with the master cards documents 
    # Which were retrieved in last space
    # Before this step we need to have 
    user_rec_cards_model = set()
    for i in docs_ids_user_query_model:
        [user_rec_cards_model.add(card) for card in master_graph[i]]
        
    # create db and return the indi
    indices_to_return = sea_complete.iloc[list(user_rec_cards_model)].index
            
    return indices_to_return
```

```python
# Create the matrix for cosine similarity query

mat_vector_query = [0] * 15
for tup in vector:
    mat_vector_query[tup[0]] = tup[1]
mat_vector_query
```

```python
# Create the matrix for jaccard similarity query

mat_vector_query_jac = [0] * 15
for tup in vector:
    mat_vector_query_jac[tup[0]] = 1 if tup[1] > 0 else 0
mat_vector_query_jac
```

```python
# Getting the matrix vector size 15 of all the seattle cards lda vectors

count = 0
mat_vectors_sea = []
for vec in vectors_cards_sea:
    if count in list_of_recs.index:
        mat_vector_sea = [0] * 15
        for tup in vec:
            mat_vector_sea[tup[0]] = tup[1]
        mat_vectors_sea.append(mat_vector_sea)
    count += 1
```

```python
# Getting the matrix vector size 15 of all the seattle cards lda vectors with jaccard

count = 0
mat_vectors_sea_jac = []
for vec in vectors_cards_sea:
    if count in list_of_recs.index:
        mat_vector_sea_jac = [0] * 15
        for tup in vec:
            mat_vector_sea_jac[tup[0]] = 1 if tup[1] > 0 else 0
        mat_vectors_sea_jac.append(mat_vector_sea_jac)
    count += 1
```

```python
len(mat_vectors_sea_jac)
```

```python
# Doing all the cosine sims between query and seattle cards

from scipy import spatial
results = []
for matvect in mat_vectors_sea:
    result = 1 - spatial.distance.cosine(mat_vector_query, matvect)
    results.append(result)
len(results)
```

```python
# Using Jaccard for similarity

results_jaccard = []
for matvect in mat_vectors_sea_jac:
    result_jaccard = 1 - spatial.distance.jaccard(mat_vector_query_jac, matvect)
    results_jaccard.append(result_jaccard)
results_jaccard
```

```python
# Adding the column to recommendations db

list_of_recs["cos_sim_query"] = results
```

```python
list_of_recs["jac_sim_query"] = results_jaccard
```
